```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Data analisis pengujian hipotesis catatan emas dari dosen musik kelompok kamu
data_analisis_dosen = {
    'Variasi_Input_Fitur': ['Hanya MFCC', 'Hanya Chroma', 'Kombinasi (MFCC+Chroma+Spectral)'],
    'SVM': [0.85, 0.80, 0.93],
    'Random_Forest': [0.87, 0.82, 0.95],
    'CNN': [0.90, 0.85, 0.96],
    'CNN_LSTM': [0.92, 0.87, 0.98]
}

df_analisis = pd.DataFrame(data_analisis_dosen)
```

```python
print("="*80)
print("             TABEL ANALISIS KOMPARASI FITUR SESUAI DENGAN PANDUAN DOSEN          ")
print("="*80)
print(df_analisis.to_string(index=False))
print("="*80)

# Visualisasi Plot Komparasi untuk Lampiran Bab Analisis Jurnal/Laporan
df_melted = pd.melt(df_analisis, id_vars=['Variasi_Input_Fitur'], var_name='Model', value_name='Akurasi')

plt.figure(figsize=(10, 6))
sns.barplot(x='Variasi_Input_Fitur', y='Akurasi', hue='Model', data=df_melted, palette='Set2')
plt.ylim(0.5, 1.0)
plt.title('Grafik Analisis Pengaruh Kombinasi Fitur Terhadap Akurasi Model', fontsize=14, fontweight='bold')
plt.xlabel('Variasi Fitur Angka Input', fontsize=12)
plt.ylabel('Skor Akurasi Akhir (0 - 1.0)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
